```python
import os
import cv2
import numpy as np
```

```python
ims_path = "/Users/zhiyumo/Downloads/Alzheimer's MRI/Test/Non/"
```

```python
ims_list = os.listdir(ims_path)
```

```python
grey_means = []
```

```python
for ims_list in ims_list:
    im = cv2.imread(ims_path+ims_list,0)
    mean = cv2.mean(im)
    grey_means.append(round(mean[0]))
```

```python
grey_means
```

```python
non_means = grey_means.copy()
```

```python
import matplotlib.pyplot as plt
import numpy as np
hist, bins = np.histogram(grey_means, bins=np.arange(min(grey_means), max(grey_means)+2))

plt.bar(bins[:-1], hist, align='center', width=0.8)
plt.xlabel('pixel value')
plt.ylabel('frequency')
plt.title('Modera train frequency histogram')

# 显示图表
plt.show()
```
