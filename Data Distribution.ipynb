```python
import os
import numpy as np
import cv2
```

```python
moderate_path = "/Users/zhiyumo/Downloads/Alzheimer's MRI/Test/Moderate/"
mild_path = "/Users/zhiyumo/Downloads/Alzheimer's MRI/Test/Mild/"
non_path = "/Users/zhiyumo/Downloads/Alzheimer's MRI/Test/Non/"
```

```python
moderate_list = os.listdir(moderate_path)
mild_list = os.listdir(mild_path)
non_list = os.listdir(non_path)
```

```python
moderate_grey_means = []
mild_grey_means = []
non_grey_means = []
```

```python
im = cv2.imread(moderate_path+"mildDem675.jpg",0)
mean = cv2.mean(im)
```

```python
for im_list in moderate_list:
    im = cv2.imread(moderate_path+im_list,0)
    mean = cv2.mean(im)
    if mean[0]/255.0 > 0.05: 
        moderate_grey_means.append(mean[0]/255.0)
```

```python
for ims_list in mild_list:
    im = cv2.imread(mild_path+ims_list,0)
    mean = cv2.mean(im)
    mild_grey_means.append(mean[0]/255.0)
```

```python
for ims_list in non_list:
    im = cv2.imread(non_path+ims_list,0)
    mean = cv2.mean(im)
    non_grey_means.append(mean[0]/255.0)
```

```python
moderate_grey_means
```

```python
import matplotlib.pyplot as plt
plt.hist(moderate_grey_means, bins=30, alpha=0.3, color='blue', label='Moderate')
plt.hist(mild_grey_means, bins=30, alpha=0.4, color='green', label='Mild')
plt.hist(non_grey_means, bins=30, alpha=0.5, color='red', label='Non')

# 
plt.title('Histogram of Train Data Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.kdeplot(moderate_grey_means, color='blue', label='Moderate', shade=True)
sns.kdeplot(mild_grey_means, color='green', label='Mild', shade=True)
sns.kdeplot(non_grey_means, color='red', label='Non', shade=True)

# Add title and labels
plt.title('Kernel Density Estimation of Test Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.stats import kstest

#perform Kolmogorov-Smirnov test
kstest(non_grey_means, 'norm')
```
